```python
import numpy as np
import matplotlib.pyplot as plt
#import function
```

```python
x = np.linspace(-10,10,20)
#Makes an array
```

```python
x
#Print variable
```

```python
m = 50
b = 17
#Defines m and b
```

```python
y = m * x + b
#Defines y as the equation given
```

```python
plt.plot(x,y,'.')
#Plots x vs y
```

```python
r = 100*np.random.randn(20,)
#Creates an array of 20 random numbers and multiplies each number by 100
```

```python
r
```

```python
plt.plot(r,'.')
#Plots the values of the r array
```

```python
y2 = m * x + b + r
```

```python
plt.plot(x,y2,'.')
#Plots x vs mx+b+r
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y2,'g.');
#Plots x+1 vs mx+b+r
```

```python
for m_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y2,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,'M = ',m_guess)
        # Makes plots of 100 values of m_guess from -100 to 100 and calculates the loss in each value or (the difference between the value*x and the value) squared
```

```python
best_loss = 100000000000
best_m = 0
best_b = 0
#Creates temporary variable values to be replaced

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:

            best_loss = loss
            best_m = m_guess
            best_b = b_guess
            #Iterates through 100 values from -100 to 100 for m_guess and b_guess, multiples the m_guess value by x and adds the b_guess value to get y_guess, then finds the loss
            #It then checks if it is the lowest value of loss yet and if it is, saves those values as the m and b values which give the smallest loss
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r.');
plt.plot(x,y2,'g.');
plt.show()

print(best_loss,best_m,best_b)
#Uses the values of m and b that had the least loss and uses them to redefine y_guess and plots it vs y2
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name):
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
    #Imports functions to run this function seperately, also imports the database used in this section
```
